```python
#Including all the required libraries:
from IPython.display import display
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Loading the data from csv file:
dataframe = pd.read_csv("Iris.csv")
```

```python
#Printing the shape of the dataset:
print("Shape of the Dataframe:", dataframe.shape)
```

```python
#Printing the column names of the dataset:
print("Column Names of the Dataframe:")
print(dataframe.columns)
```

```python
#Printing the first 3 rows from the csv file:
display(dataframe.head(3))
```

```python
#Using .info() for summary statistics
#to get information regarding data columns from the csv file:
display(dataframe.info())
```

```python
#Using .describe() for summary statistics
#to get the count, mean, std, min and max of the data columns in the csv file:
display(dataframe.describe())
```

```python
#Creating a scatterplot to show relationship between sepal and petal lengths:
plt.figure()
sns.scatterplot(x="SepalLengthCm", y="PetalLengthCm", data=dataframe)
plt.title("Scatterplot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.show()
```

```python
#Creating a histogram to show value distributions:
plt.figure()
sns.histplot(data=dataframe, x = "SepalLengthCm")
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
#Creating a boxplot to identify outliers:
plt.figure()
sns.boxplot(data = dataframe, x = "Species", y = "SepalLengthCm")
plt.title("Box plot of Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length")
plt.show()
```
